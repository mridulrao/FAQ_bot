```python
def get_knowledge():
    with open("/Users/kawaii/Desktop/faq_bot/faq_bot/process_query/knowledge.txt") as file:
        knowledge = file.read()
    
    file.close()
    return knowledge
```

```python
knowledge = get_knowledge()
```

# Create Intent Words

```python
import nltk 
from nltk.corpus import stopwords
sw_nltk = stopwords.words('english')
```

```python
words = [word for word in knowledge.split() if word.lower() not in sw_nltk]
new_knowledge = " ".join(words)
print(new_knowledge)
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
knowledge_lemmatize = lemmatizer.lemmatize(knowledge)
```

```python
words = [word for word in knowledge_lemmatize.split() if word.lower() not in sw_nltk]
new_knowledge_lemmatize = " ".join(words)
print(new_knowledge_lemmatize)
```

```python
words_1 = new_knowledge.split()
words_2 = new_knowledge_lemmatize.split()
```

```python
words_1
```

```python
words_2
```

```python
bag_of_words = words_1

print(len(bag_of_words))

for word in words_2:
    if word not in bag_of_words:
        bag_of_words.append(word)
        
print(len(bag_of_words))
```
